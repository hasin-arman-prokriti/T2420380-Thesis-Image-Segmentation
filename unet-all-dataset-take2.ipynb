```python
import os
import cv2
import numpy as np
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt

# Base directory containing all datasets
base_path = "DataSet"

# Load and preprocess data from multiple sources
def load_data(base_path, image_size=(256, 256)):
    images, masks = [], []
    # Iterate through each dataset folder
    for dataset_name in os.listdir(base_path):
        dataset_path = os.path.join(base_path, dataset_name)
        image_path = os.path.join(dataset_path, "images")
        mask_path = os.path.join(dataset_path, "masks")
        
        # Ensure both image and mask directories exist
        if not os.path.exists(image_path) or not os.path.exists(mask_path):
            print(f"Skipping dataset: {dataset_name} (Missing 'images' or 'masks')")
            continue
        
        # Iterate through files in the image directory
        for file_name in os.listdir(image_path):
            try:
                # Load image and corresponding mask
                img = cv2.imread(os.path.join(image_path, file_name))
                mask = cv2.imread(os.path.join(mask_path, file_name), 0)  # Load mask as grayscale
                
                # Skip if loading failed
                if img is None or mask is None:
                    print(f"Skipping file: {file_name} in {dataset_name}")
                    continue
                
                # Resize and normalize
                img = cv2.resize(img, image_size) / 255.0
                mask = cv2.resize(mask, image_size) / 255.0
                
                images.append(img)
                masks.append(mask)
            except Exception as e:
                print(f"Error processing file {file_name} in {dataset_name}: {e}")
    
    images = np.array(images, dtype=np.float32)
    masks = np.expand_dims(np.array(masks, dtype=np.float32), axis=-1)  # Add channel dimension
    return images, masks

# Load dataset
images, masks = load_data(base_path)

# Split into training and validation sets
X_train, X_val, y_train, y_val = train_test_split(images, masks, test_size=0.2, random_state=42)

# Display a sample
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.imshow(X_train[0])
plt.title("Image")
plt.subplot(1, 2, 2)
plt.imshow(y_train[0].squeeze(), cmap="gray")
plt.title("Mask")
plt.show()
```

```python
from tensorflow.keras import layers, models

def unet(input_shape=(256, 256, 3)):
    inputs = layers.Input(input_shape)
    
    # Encoder
    c1 = layers.Conv2D(64, (3, 3), activation='relu', padding='same')(inputs)
    c1 = layers.Conv2D(64, (3, 3), activation='relu', padding='same')(c1)
    p1 = layers.MaxPooling2D((2, 2))(c1)

    c2 = layers.Conv2D(128, (3, 3), activation='relu', padding='same')(p1)
    c2 = layers.Conv2D(128, (3, 3), activation='relu', padding='same')(c2)
    p2 = layers.MaxPooling2D((2, 2))(c2)

    # Bottleneck
    c3 = layers.Conv2D(256, (3, 3), activation='relu', padding='same')(p2)
    c3 = layers.Conv2D(256, (3, 3), activation='relu', padding='same')(c3)

    # Decoder
    u1 = layers.UpSampling2D((2, 2))(c3)
    u1 = layers.Concatenate()([u1, c2])
    c4 = layers.Conv2D(128, (3, 3), activation='relu', padding='same')(u1)
    c4 = layers.Conv2D(128, (3, 3), activation='relu', padding='same')(c4)

    u2 = layers.UpSampling2D((2, 2))(c4)
    u2 = layers.Concatenate()([u2, c1])
    c5 = layers.Conv2D(64, (3, 3), activation='relu', padding='same')(u2)
    c5 = layers.Conv2D(64, (3, 3), activation='relu', padding='same')(c5)

    outputs = layers.Conv2D(1, (1, 1), activation='sigmoid')(c5)

    return models.Model(inputs, outputs)

# Create the model
model = unet()
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
model.summary()
```

```python
history = model.fit(
    X_train, y_train,
    validation_data=(X_val, y_val),
    batch_size=24,
    epochs=10
)
```

```python
# Evaluate the model
loss, accuracy = model.evaluate(X_val, y_val)
print(f"Validation Loss: {loss}, Accuracy: {accuracy}")

# Predict and visualize
predicted_mask = model.predict(np.expand_dims(X_val[0], axis=0))[0]
plt.figure(figsize=(10, 5))
plt.subplot(1, 3, 1)
plt.imshow(X_val[0])
plt.title("Input Image")
plt.subplot(1, 3, 2)
plt.imshow(y_val[0].squeeze(), cmap="gray")
plt.title("True Mask")
plt.subplot(1, 3, 3)
plt.imshow(predicted_mask.squeeze(), cmap="gray")
plt.title("Predicted Mask")
plt.show()
```

```python
import os
import cv2
import numpy as np
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt

# Base directory containing all datasets
base_path = "DataSet"

# Load and preprocess data from multiple sources
def load_data(base_path, image_size=(256, 256)):
    images, masks = [], []
    # Iterate through each dataset folder
    for dataset_name in os.listdir(base_path):
        dataset_path = os.path.join(base_path, dataset_name)
        image_path = os.path.join(dataset_path, "images")
        mask_path = os.path.join(dataset_path, "masks")
        
        # Ensure both image and mask directories exist
        if not os.path.exists(image_path) or not os.path.exists(mask_path):
            print(f"Skipping dataset: {dataset_name} (Missing 'images' or 'masks')")
            continue
        
        # Iterate through files in the image directory
        for file_name in os.listdir(image_path):
            try:
                # Load image and corresponding mask
                img = cv2.imread(os.path.join(image_path, file_name))
                mask = cv2.imread(os.path.join(mask_path, file_name), 0)  # Load mask as grayscale
                
                # Skip if loading failed
                if img is None or mask is None:
                    print(f"Skipping file: {file_name} in {dataset_name}")
                    continue
                
                # Resize and normalize
                img = cv2.resize(img, image_size) / 255.0
                mask = cv2.resize(mask, image_size) / 255.0
                
                images.append(img)
                masks.append(mask)
            except Exception as e:
                print(f"Error processing file {file_name} in {dataset_name}: {e}")
    
    images = np.array(images, dtype=np.float32)
    masks = np.expand_dims(np.array(masks, dtype=np.float32), axis=-1)  # Add channel dimension
    return images, masks

# Load dataset
images, masks = load_data(base_path)

# Split into training and validation sets
X_train, X_val, y_train, y_val = train_test_split(images, masks, test_size=0.2, random_state=42)

# Display a sample
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.imshow(X_train[0])
plt.title("Image")
plt.subplot(1, 2, 2)
plt.imshow(y_train[0].squeeze(), cmap="gray")
plt.title("Mask")
plt.show()
```

```python
#tensorflow
import tensorflow as tf
from tensorflow.keras import layers, Model

def unet_model(input_shape=(256, 256, 3)):
    inputs = tf.keras.Input(input_shape)

    # Downsample
    c1 = layers.Conv2D(64, (3, 3), activation='relu', padding='same')(inputs)
    c1 = layers.Conv2D(64, (3, 3), activation='relu', padding='same')(c1)
    p1 = layers.MaxPooling2D((2, 2))(c1)

    c2 = layers.Conv2D(128, (3, 3), activation='relu', padding='same')(p1)
    c2 = layers.Conv2D(128, (3, 3), activation='relu', padding='same')(c2)
    p2 = layers.MaxPooling2D((2, 2))(c2)

    c3 = layers.Conv2D(256, (3, 3), activation='relu', padding='same')(p2)
    c3 = layers.Conv2D(256, (3, 3), activation='relu', padding='same')(c3)
    p3 = layers.MaxPooling2D((2, 2))(c3)

    # Base
    base = layers.Conv2D(512, (3, 3), activation='relu', padding='same')(p3)
    base = layers.Conv2D(512, (3, 3), activation='relu', padding='same')(base)

    # Upsample
    u1 = layers.Conv2DTranspose(256, (2, 2), strides=(2, 2), padding='same')(base)
    u1 = layers.concatenate([u1, c3])
    c4 = layers.Conv2D(256, (3, 3), activation='relu', padding='same')(u1)
    c4 = layers.Conv2D(256, (3, 3), activation='relu', padding='same')(c4)

    u2 = layers.Conv2DTranspose(128, (2, 2), strides=(2, 2), padding='same')(c4)
    u2 = layers.concatenate([u2, c2])
    c5 = layers.Conv2D(128, (3, 3), activation='relu', padding='same')(u2)
    c5 = layers.Conv2D(128, (3, 3), activation='relu', padding='same')(c5)

    u3 = layers.Conv2DTranspose(64, (2, 2), strides=(2, 2), padding='same')(c5)
    u3 = layers.concatenate([u3, c1])
    c6 = layers.Conv2D(64, (3, 3), activation='relu', padding='same')(u3)
    c6 = layers.Conv2D(64, (3, 3), activation='relu', padding='same')(c6)

    # Output Layer
    outputs = layers.Conv2D(1, (1, 1), activation='sigmoid')(c6)

    # Model
    model = Model(inputs, outputs)
    return model

# Instantiate the model
unet = unet_model()
unet.summary()
```

```python
import tensorflow as tf
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.losses import BinaryCrossentropy
from tensorflow.keras.metrics import MeanIoU
from tensorflow.keras.callbacks import ModelCheckpoint, EarlyStopping
import matplotlib.pyplot as plt

# Define parameters
EPOCHS = 20
BATCH_SIZE = 16
LEARNING_RATE = 1e-4

# Prepare dataset
train_dataset = tf.data.Dataset.from_tensor_slices((X_train, y_train))
train_dataset = train_dataset.shuffle(1000).batch(BATCH_SIZE).prefetch(tf.data.AUTOTUNE)

val_dataset = tf.data.Dataset.from_tensor_slices((X_val, y_val))
val_dataset = val_dataset.batch(BATCH_SIZE).prefetch(tf.data.AUTOTUNE)

# Compile the model
unet = unet_model(input_shape=(256, 256, 3))
unet.compile(
    optimizer=Adam(learning_rate=LEARNING_RATE),
    loss=BinaryCrossentropy(),
    metrics=[MeanIoU(num_classes=2)]
)

# Callbacks
checkpoint = ModelCheckpoint(
    "unet_best_model.keras", 
    save_best_only=True, 
    monitor="val_loss", 
    mode="min", 
    verbose=1
)
early_stopping = EarlyStopping(
    monitor="val_loss", 
    patience=5, 
    mode="min", 
    verbose=1
)

# Train the model
history = unet.fit(
    train_dataset,
    validation_data=val_dataset,
    epochs=EPOCHS,
    callbacks=[checkpoint, early_stopping]
)

# Plot training and validation loss
plt.figure(figsize=(10, 5))
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Loss Curve')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.show()

# Plot Mean IoU
plt.figure(figsize=(10, 5))
plt.plot(history.history['mean_io_u'], label='Training IoU')
plt.plot(history.history['val_mean_io_u'], label='Validation IoU')
plt.title('Mean IoU Curve')
plt.xlabel('Epochs')
plt.ylabel('Mean IoU')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Function to make a prediction and visualize
def predict_and_display(image, true_mask, model):
    # Preprocess the input image
    image_resized = cv2.resize(image, (256, 256)) / 255.0
    image_resized = np.expand_dims(image_resized, axis=0)  # Add batch dimension

    # Predict the mask
    predicted_mask = model.predict(image_resized)[0]
    predicted_mask = (predicted_mask > 0.5).astype(np.uint8)  # Binarize the mask

    # Display the results
    plt.figure(figsize=(15, 5))
    plt.subplot(1, 3, 1)
    plt.imshow(image)
    plt.title("Original Image")
    plt.axis("off")

    plt.subplot(1, 3, 2)
    plt.imshow(true_mask.squeeze(), cmap="gray")
    plt.title("True Mask")
    plt.axis("off")

    plt.subplot(1, 3, 3)
    plt.imshow(predicted_mask.squeeze(), cmap="gray")
    plt.title("Predicted Mask")
    plt.axis("off")

    plt.show()

# Select a random image and mask from the validation dataset
idx = np.random.randint(0, len(X_val))
sample_image = X_val[idx]
sample_mask = y_val[idx]

# Predict and display
predict_and_display(sample_image, sample_mask, unet)

# Debugging: Check predicted mask values
predicted_mask = model.predict(image_resized)[0]
print("Predicted Mask Min:", predicted_mask.min())
print("Predicted Mask Max:", predicted_mask.max())
```

```python
# Evaluate the model
loss, accuracy = model.evaluate(X_val, y_val)
print(f"Validation Loss: {loss}, Accuracy: {accuracy}")

# Predict and visualize
predicted_mask = model.predict(np.expand_dims(X_val[0], axis=0))[0]
plt.figure(figsize=(10, 5))
plt.subplot(1, 3, 1)
plt.imshow(X_val[0])
plt.title("Input Image")
plt.subplot(1, 3, 2)
plt.imshow(y_val[0].squeeze(), cmap="gray")
plt.title("True Mask")
plt.subplot(1, 3, 3)
plt.imshow(predicted_mask.squeeze(), cmap="gray")
plt.title("Predicted Mask")
plt.show()
```
